# 06. Task step 6
## Create a spending flag for each user based on the average price across all their orders using the following criteria: 
    1. If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
    2. If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# creating 'average_price' column using transform()

ords_prods_merged['average_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# check if 'average_price' column was created successfully

ords_prods_merged.head()
```

```python
# creating a 'spending_flag' flag

ords_prods_merged.loc[ords_prods_merged['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# creating a 'spending_flag' flag

ords_prods_merged.loc[ords_prods_merged['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# check if flag was created properly

ords_prods_merged.head()
```

```python
# check the frequency

ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

# 07. Task step 7

## Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
    1. If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
    2. If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
    3. If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# creating “median_prior_orders” column using transform()

ords_prods_merged['median_prior_orders'] = ords_prods_merged.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
# check new column was added

ords_prods_merged.head()
```

```python
# creating a 'order_frequency_flag' flag

ords_prods_merged.loc[ords_prods_merged['median_prior_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# creating a 'order_frequency_flag' flag

ords_prods_merged.loc[(ords_prods_merged['median_prior_orders'] > 10) & (ords_prods_merged['median_prior_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
# creating a 'order_frequency_flag' flag

ords_prods_merged.loc[ords_prods_merged['median_prior_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# check if 'order_frequency_flag' was created successfully

ords_prods_merged.head()
```

```python
# check the frequency

ords_prods_merged['order_frequency_flag'].value_counts(dropna = False)
```
