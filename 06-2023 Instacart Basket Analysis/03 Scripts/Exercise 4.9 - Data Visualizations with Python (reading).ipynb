# 07. Sampling data

```python
# seed created random but reproducible pattern
# want a 70/30 split so value is 0.7
# random creates the random values
# rand ensures the numbers created by random follow a uniform distribution

np.random.seed(4)
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
# dev creates a list of boolean values based on whether the actual value is <= 0.7

dev
```

```python
# store 70% of the sample in the dataframe big

big = ords_prods_merge[dev]
```

```python
# store 30% of the sample in the dataframe small

small = ords_prods_merge[~dev]
```

```python
# check if big + small = ords_prods_merge

len(ords_prods_merge)
```

```python
len(big) + len(small)
```

```python
# limit columns to those necessary for chart

df_2 = small[['orders_day_of_week','prices']]
```

```python
# creating line chart

line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# check if line chart is representative of entire population

df_4 = ords_prods_merge[['orders_day_of_week', 'prices']]
```

```python
line_2 = sns.lineplot(data = df_4, x = 'orders_day_of_week',y = 'prices')
```
