# 05. Task step 5
## Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day. (Hint: To check this, you need to use an accurate sample for your line chart!)

```python
# create random sample of data with 70/30 split

np.random.seed(4)
dev = np.random.rand(len(ords_prods_cust_merge)) <= 0.7
```

```python
# store 70% of the sample in the dataframe big

big = ords_prods_cust_merge[dev]
```

```python
# store 30% of the sample in the dataframe small

small = ords_prods_cust_merge[~dev]
```

```python
# check if length of big + small matches entire dataframe

len(ords_prods_cust_merge)
```

```python
len(big) + len(small)
```

```python
# limit columns to those necessary for chart

df = small[['order_hour_of_day','prices']]
```

```python
# creating line chart

line = sns.lineplot(data = df, x = 'order_hour_of_day',y = 'prices')
```

## There is a spike in the morning (hours 2-4) where more expensive produce is bought followed by a dip until hour 12, then the price levels out for the rest of the day.

```python
# exporting line chart

line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_orders_hour_of_day_prices.png'))
```
